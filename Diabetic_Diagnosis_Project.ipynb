```
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
from scipy.stats import chi2_contingency, f_oneway

print("Libraries Imported Successfully")
```

```
# Create Sample Dataset

np.random.seed(42)

n = 100

df = pd.DataFrame({
    "record_id": [f"P{i}" for i in range(1, n+1)],
    "age": np.random.randint(20, 70, n),
    "weight": np.random.randint(45, 100, n),
    "gender": np.random.choice(["Male", "Female"], n),
    "region": np.random.choice(["North", "South", "East", "West"], n),
    "smoking_status": np.random.choice(["Smoker", "Non-Smoker"], n),
    "exercise_frequency": np.random.choice(["Daily", "Weekly", "Rarely", "Never"], n),
    "bmi": np.round(np.random.uniform(18, 35, n), 2),
    "blood_pressure": np.round(np.random.uniform(80, 160, n), 2),
    "glucose_level": np.round(np.random.uniform(70, 220, n), 2),
    "cholesterol_level": np.round(np.random.uniform(150, 300, n), 2),
    "diabetes": np.random.choice([True, False], n),
    "hypertension": np.random.choice([True, False], n)
})

df.head()
```

## Dataset Information

```
# Dataset Information
df.info()
```

```
# Statistical Summary
df.describe()
```

## Descriptive Statistics

```
# Mean, Median, Mode

print("Mean:\n")
print(df[['age','weight','bmi','glucose_level']].mean())

print("\nMedian:\n")
print(df[['age','weight','bmi','glucose_level']].median())

print("\nMode:\n")
print(df[['age','weight','bmi','glucose_level']].mode().iloc[0])
```

```
# Variance and Standard Deviation

print("Variance:\n")
print(df[['age','weight','bmi','glucose_level']].var())

print("\nStandard Deviation:\n")
print(df[['age','weight','bmi','glucose_level']].std())
```

## Confidence Interval

```
# 95% Confidence Interval for BMI

bmi_mean = df['bmi'].mean()
bmi_std = df['bmi'].std()
n = len(df)

margin_error = 1.96 * (bmi_std / np.sqrt(n))

lower = bmi_mean - margin_error
upper = bmi_mean + margin_error

print("95% Confidence Interval for BMI:")
print((lower, upper))
```

## Correlation and Covariance

```
# Correlation Matrix

corr = df[['age','weight','bmi','glucose_level','blood_pressure']].corr()

corr
```

```
# Heatmap

plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```
# Covariance Matrix

cov = df[['age','weight','bmi','glucose_level']].cov()

cov
```

## Chi-Square Test

```
# Chi-Square Test between Smoking Status and Diabetes

contingency_table = pd.crosstab(df['smoking_status'], df['diabetes'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Value:", chi2)
print("P-Value:", p)

if p < 0.05:
    print("There is a significant relationship.")
else:
    print("No significant relationship.")
```

## ANOVA Test

```
# ANOVA Test for BMI based on Exercise Frequency

daily = df[df['exercise_frequency'] == 'Daily']['bmi']
weekly = df[df['exercise_frequency'] == 'Weekly']['bmi']
rarely = df[df['exercise_frequency'] == 'Rarely']['bmi']
never = df[df['exercise_frequency'] == 'Never']['bmi']

f_stat, p_value = f_oneway(daily, weekly, rarely, never)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Significant difference exists.")
else:
    print("No significant difference.")
```

## Data Visualization

```
# Histogram for Glucose Level

plt.figure(figsize=(6,4))
plt.hist(df['glucose_level'], bins=10)
plt.title("Glucose Level Distribution")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()
```

```
# Scatter Plot between BMI and Glucose Level

plt.figure(figsize=(6,4))
plt.scatter(df['bmi'], df['glucose_level'])
plt.xlabel("BMI")
plt.ylabel("Glucose Level")
plt.title("BMI vs Glucose Level")
plt.show()
```
